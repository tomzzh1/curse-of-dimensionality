**The curse of dimensionality**

> As the number of dimensionality increases, the number of observation needed to establish significant relationship between variables grow significantly. 

> if we simulate a random stock for 20 days, 0 represents down and 1 represents up, then we flip a coin enough to mimic stock's behavior. If we flip a coin enough time, we will find the "perfect coin" to predict stock price.

```python
import numpy as np
import pandas as pd


stock = np.random.randint(low = 0, high = 2, size = 10)
stock
```

```python
coins = []
coin = np.random.randint(low = 0, high = 2, size = 10)
data = pd.DataFrame()


while (coin != stock).any():
    
    data['Coin ' + str(len(list(data)))] = coin 
    
    coin = np.random.randint(low = 0, high = 2, size = 10)
    

data['Coin ' + str(len(list(data)))] = coin
data['stock'] = stock
```
